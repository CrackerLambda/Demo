## Ưu điểm:

Python là một **ngôn ngữ hướng đối tượng** có thể viết rất cơ bản. Mô hình class của python cũng hỗ trợ các khái niệm nâng cao như tính đa hình, nạp chồng toán tử và nhiều giá trị kế thừa.


Có thể **quản lý bộ nhớ tự động** Python tự động phân bổ các đối tượng và thu hồi (“collection garbage”) chúng khi chúng không còn được sử dụng nữa và hầu hết có thể phát triển và thu nhỏ theo yêu cầu. 

**Hỗ trợ lập trình trong quy mô lớn** để xây dựng các hệ thống lớn hơn, Python bao gồm các công cụ như module, lớp và ngoại lệ. Các công cụ này cho phép bạn tổ chức hệ thống thành các thành phần, sử dụng OOP để tái sử dụng và tùy chỉnh mã, đồng thời xử lý các sự kiện và lỗi một cách hợp lý.

Python cung cấp các **cấu trúc dữ liệu** thường được sử dụng như danh sách, từ điển và chuỗi như là các phần lõi của ngôn ngữ; như bạn sẽ thấy, chúng đều linh hoạt và dễ sử dụng.

**Rất dễ sử dụng** để chạy một chương trình Python, bạn chỉ cần gõ nó và chạy nó. Không có các bước biên dịch và liên kết, trung gian, giống như các ngôn ngữ như C hoặc C ++. Python thực thi các chương trình ngay lập tức

**Rất dễ học** so với các ngôn ngữ lập trình khác, ngôn ngữ Python cốt lõi rất dễ học.

Python đi kèm với một **bộ sưu tập lớn các thư viện hỗ trợ** các chức năng được tạo sẵn được gọi là thư viện tiêu chuẩn. Thư viện này hỗ trợ một loạt các nhiệm vụ vẽ đồ thị chuyên nghiệp ở cấp ứng dụng.Ví dụ, phần mở rộng NumPy đã được mô tả là một phần mềm miễn phí và mạnh
mẽ hơn tương đương với hệ thống lập trình số Matlab.

## Biên dịch mã bytecode

Khi bạn thực thi một chương trình, Python đầu tiên biên dịch mã nguồn của bạn (các câu lệnh trong tệp của bạn) thành một định dạng được gọi là mã byte thành cách trình bày **mã cấp thấp hơn**, độc lập với nền tảng của mã nguồn của bạn. Về cơ bản, Python dịch từng câu lệnh nguồn của bạn thành một nhóm hướng dẫn mã byte bằng cách phân tách chúng thành các bước riêng lẻ. Việc dịch mã byte này được thực hiện để tăng tốc độ thực thi — mã byte có thể chạy nhanh hơn nhiều so với các câu lệnh mã nguồn ban đầu trong tệp văn bản của bạn.

Bạn sẽ nhận thấy rằng đoạn trước nói rằng điều này gần như hoàn toàn bị che giấu với bạn. Nếu
quy trình Python có quyền ghi trên máy của bạn, nó sẽ lưu trữ mã byte của các chương trình của
bạn trong các tệp kết thúc bằng phần mở rộng .pyc (“.pyc” có nghĩa là nguồn “.py” được biên
dịch).

````python
import dis
def printString():
    print("Hello world!")
dis.dis(printString)
````

## Máy ảo Python (Python Virtual Machine)

Sau khi chương trình của bạn đã được biên dịch thành mã byte (hoặc mã byte đã được tải từ các
tệp .pyc hiện có ), nó sẽ được chuyển đi để thực thi đến một thứ thường được gọi là máy ảo Python
(PVM, để bạn có xu hướng viết tắt nhiều hơn. ). Trên thực tế, PVM chỉ là một vòng lặp lớn lặp đi lặp lại lần lượt các hướng dẫn mã byte của bạn để thực hiện các hoạt động của chúng. PVM là công cụ thời gian chạy của Python; nó luôn hiện diện như một phần của hệ thống Python và nó là thành phần thực sự chạy các tập lệnh của bạn. Về mặt kỹ thuật, đó chỉ là bước cuối cùng của cái được gọi là **“Trình thông dịch Python”**.
