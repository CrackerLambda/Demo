## 1. Lịch sử ngôn ngữ lập trình Python

Python ra đời vào vào tháng 12 năm 1989 do Guido van Rossum người Hà Lan sáng lập ra.

- Version Python 1.0 ra đời vào năm 1991

- Version Python 2.0 ra đời vào năm 2000

- Version Python 3.0 ra đời năm 2008 và được phát triển đến ngày nay

Python là thuộc dạng ngôn ngữ bậc cao, khác với C,C++ cần có compiler trình biên dịch để dịch sang mã máy thì python xài trình thông dịch interpreter nên khi viết chỉ cần xuống dòng, thụt lụt lề để chương trình hiểu dịch ra từng dòng lệnh.

## 2. Ứng dụng của Python

- **Rèn luyện tư duy toán học**: Sử dụng hỗ trợ tính toán với ma trận, đạo hàm, tối ưu hóa, ...nhờ các thư viện có sẵn như numpy, scipy.
- **Trực quan hóa dữ liệu**: Nhờ các thư viện như pandas, matplotlib, seaborn mà ta có thể phân tích dữ liệu bằng biểu đồ như cột, tròn, thống kê từ đó ta có cái nhìn sâu sắc về dữ liệu.
- **Machine learning**: Ngày nay với cuộc cách mạng 4.0 nỗi lên như vũ bão đòi hỏi người ta càng ứng dụng các mô hình thuật toán đã xây dựng trước đó vào triển khai thực tế như bài toán chuyển giọng nói thành văn bản sử dụng RNN, nhận dạng khuôn mặt của con người bằng CNN, phân loại nhóm khách hàng của một cửa hàng, trung tâm mua sắm, ...
- **Lập trình game**: Thư viện pygame
- **Thiết kế trang web**: Sử dụng framework django
